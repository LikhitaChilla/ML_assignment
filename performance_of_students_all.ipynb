```python
x = df.drop('Exam_Score',axis=1)
y = df['Exam_Score']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import pickle
from sklearn.preprocessing import LabelEncoder

# Features and their encoders
encoders = {
    'Parental_Involvement': LabelEncoder(),
    'Access_to_Resources': LabelEncoder(),
    'Extracurricular_Activities': LabelEncoder(),
    'Motivation_Level': LabelEncoder(),
    'Internet_Access': LabelEncoder(),
    'Family_Income': LabelEncoder(),
    'Teacher_Quality': LabelEncoder(),
    'School_Type': LabelEncoder(),
    'Peer_Influence': LabelEncoder(),
    'Learning_Disabilities': LabelEncoder(),
    'Parental_Education_Level': LabelEncoder(),
    'Distance_from_Home': LabelEncoder(),
    'Gender': LabelEncoder()
}

# Example data for training (replace this with your actual training data)

# Dictionary to store the transformed data
transformed_data = {}

# Fit, transform, and pickle each encoder
for feature, encoder in encoders.items():
    
    # Fit and transform the training data
    transformed_data[feature] = encoder.fit_transform(x_train[feature])
    x_train[feature]= encoder.fit_transform(x_train[feature])
    x_test[feature]=encoder.transform(x_test[feature])
 
    # Save the encoder
    with open(f'{feature}_encoder.pkl', 'wb') as f:
        pickle.dump(encoder, f)

print("All encoders have been fit, transformed, and pickled successfully!")
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
mse=mean_absolute_error(y_pred,y_test)
mse
```

```python
r2_Score=r2_score(y_test,y_pred)
r2_Score
```

```python
from sklearn.preprocessing import LabelEncoder, PolynomialFeatures, StandardScaler
```

```python
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
lr.fit(x_train_poly, y_train)
y_pred = lr.predict(x_test_poly)
r2_score_poly = r2_score(y_test, y_pred)
r2_score_poly
```
